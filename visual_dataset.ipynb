```python
import os
import numpy as np
import pickle as pk
import pandas as pd
import matplotlib.pyplot as plt

pth = './'

dataset = 'SMD'

with open(pth + dataset + '/' + dataset + '.pk', 'rb') as file:
    data = pk.load(file)
    
# 从读取的数据中获取特定键对应的值
train_data = data['train_data']
test_data = data['test_data']
test_label = data['test_label']

train_x = np.arange(0, train_data.shape[0])
test_x = np.arange(0, test_data.shape[0])

print("Data loading complete")
```

```python
# train_data
plt.figure(figsize=(18, train_data.shape[1] * 3))

for i in range(train_data.shape[1]):
    plt.subplot(train_data.shape[1],1,i + 1)
    plt.plot(train_x, train_data[:, i])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'{dataset} - train - {i}')
    
plt.tight_layout()
plt.show()
    
# test_data
plt.figure(figsize=(18, test_data.shape[1] * 3))

for i in range(test_data.shape[1]):
    plt.subplot(test_data.shape[1],1,i + 1)
    plt.plot(test_x, test_data[:, i])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'{dataset} - test - {i}')
    
plt.tight_layout()
plt.show()

# test_label
plt.figure(figsize=(18, 3))

plt.subplot(1, 1, 1)
plt.plot(test_x, test_label)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'{dataset} - test_label')

plt.tight_layout()
plt.show()
            
```
